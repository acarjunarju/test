```python
# Implement A O* Search Algorithm in Python

class Node:
    def __init__(self, name, heuristic=0):
        self.name = name
        self.heuristic = heuristic
        self.children = []   # List of AND/OR child groups

    def add_children(self, child_group):
        self.children.append(child_group)


def ao_star(node):
    # If node has no children, return its heuristic value
    if not node.children:
        return node.heuristic

    minimum_cost = float('inf')
    best_group = None

    # Check all child groups
    for group in node.children:
        cost = 0

        for child in group:
            cost += ao_star(child)

        if cost < minimum_cost:
            minimum_cost = cost
            best_group = group

    print("Best path from", node.name, ":", [n.name for n in best_group],
          "with cost =", minimum_cost)

    return minimum_cost + node.heuristic


# Creating nodes
A = Node('A', 1)
B = Node('B', 3)
C = Node('C', 2)
D = Node('D', 4)
E = Node('E', 6)
F = Node('F', 5)
G = Node('G', 0)

# Defining AND-OR graph
A.add_children([B, C])   # AND relationship
A.add_children([D])      # OR relationship

B.add_children([E])
C.add_children([F, G])

result = ao_star(A)

print("\nOptimal Cost =", result)
```

```python
def negation(literal):
    """Return the negation of a literal."""
    literal = literal.strip()

    if literal.startswith("~"):
        return literal[1:]
    else:
        return "~" + literal


def resolve(clause1, clause2):
    """Perform resolution between two clauses."""
    resolvents = []

    for literal in clause1:

        complement = negation(literal)

        if complement in clause2:

            new_clause = set(clause1 + clause2)

            new_clause.discard(literal)
            new_clause.discard(complement)

            tautology = False
            for lit in new_clause:
                if negation(lit) in new_clause:
                    tautology = True
                    break

            if not tautology:
                resolvent = sorted(list(new_clause))

                if resolvent not in resolvents:
                    resolvents.append(resolvent)

    return resolvents


def resolution(kb):
    """Determine satisfiability using resolution."""

    clauses = []
    for clause in kb:

        literals = [lit.strip() for lit in clause.split("||")]

        literals = sorted(list(set(literals)))

        tautology = False
        for lit in literals:
            if negation(lit) in literals:
                tautology = True
                break

        if not tautology and literals not in clauses:
            clauses.append(literals)

    while True:

        new_clauses = []

        n = len(clauses)

        for i in range(n):
            for j in range(i + 1, n):

                resolvents = resolve(clauses[i], clauses[j])

                for resolvent in resolvents:

                    if resolvent == []:
                        return "Unsatisfiable"

                    if resolvent not in clauses and resolvent not in new_clauses:
                        new_clauses.append(resolvent)

        if not new_clauses:
            return "Satisfiable"

        clauses.extend(new_clauses)


print("Enter number of clauses:")
n = int(input())

knowledge_base = []

print("Enter clauses using || symbol")
print("Example: P||Q||~R")

for i in range(n):
    clause = input(f"Enter clause {i + 1}: ")
    knowledge_base.append(clause)

result = resolution(knowledge_base)

print("\nResult:")
print(result)
```
